```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name_1 = []

for i in soup.find_all('a',class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"):
    name_1.append(i.text)
    
name_1
```

```python
price_1 = []

for i in soup.find_all('span',class_="a-price-whole"):
    price_1.append(i.text)

price_1
```

```python
del_1 =[]

for i in soup.find_all('span',class_="a-color-base a-text-bold"):
    del_1.append(i.text)
    
del_1
```

```python
url_1 = [] 

for i in soup.find_all('a',class_="a-link-normal s-no-outline"):
    url_1.append(i.text)
    
url_1
```

```python
import pandas as pd

df = pd.DataFrame({'Product Name & Description':name_1,'Price':price_1})
df
```
